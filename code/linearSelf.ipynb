```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
dataset = pd.read_csv('/Users/hp/Desktop/group/YearPredictionMSD.txt', header = None) 
dataset
```

```python
X = dataset.iloc[:1000, 1:].as_matrix()
y = dataset.iloc[:1000, 0].as_matrix()
```

```python
def compute_cost(X, y, params):
    n_samples = len(y)
    h = X @ params
    return (1/n_samples)*np.sum((h-y)**2)
```

```python
def gradient_descent(X, y, params, learning_rate, n_iters):
    n_samples = len(y)
    J_history = np.zeros((n_iters,1))
    for i in range(n_iters):
        params = params - (learning_rate/n_samples) * X.T @ (X @ params - y) 
        J_history[i] = compute_cost(X, y, params)
    return (J_history, params)
```

```python
n_samples = len(y)
mu = np.mean(X, 0)
sigma = np.std(X, 0)
X = (X-mu) / sigma
X = np.hstack((np.ones((n_samples,1)),X))
n_features = np.size(X,1)
params = np.zeros((n_features,1))
```

```python
n_epoch = 100
learning_rate = 0.1
(J_history, optimal_params) = gradient_descent(X, y, params, learning_rate, n_epoch)
for i in range(0,n_epoch):
    print("epoch",i+1,"/100 mse is: ", J_history[i])
plt.plot(range(len(J_history)), J_history, 'r')
plt.title("Convergence Graph of Cost Function")
plt.xlabel("Number of Iterations")
plt.ylabel("Cost")
plt.show()
```
