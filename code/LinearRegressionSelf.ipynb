```python
class LinearRegression:
    #Define all parameters
    def __init__(self, X, y, alpha=0.03, n_iter=1500):
        self.alpha = alpha
        self.n_iter = n_iter
        self.n_samples = len(y)
        self.n_features = np.size(X, 1)
        self.X = np.hstack((np.ones((self.n_samples, 1)), (X - np.mean(X, 0)) / np.std(X, 0)))
        self.y = y[:, np.newaxis]
        self.params = np.zeros((self.n_features + 1, 1))
        self.coef_ = None
        self.intercept_ = None
    
    #Linear Regression model
    def fit(self):
        for i in range(self.n_iter):
            self.params = self.params - (self.alpha/self.n_samples) * \
            self.X.T @ (self.X @ self.params - self.y)
            self.intercept_ = self.params[0]
            self.coef_ = self.params[1:]
        return self
    
    #define predict function with the help of the model
    def predict(self, X):
        n_samples = np.size(X, 0)
        y = np.hstack((np.ones((n_samples, 1)), (X-np.mean(X, 0)) / np.std(X, 0))) @ self.params
        return y

    def get_params(self):
        return self.params
```

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
#load the dataset
dataset = pd.read_csv('/Users/hp/Desktop/group/YearPredictionMSD.txt', header = None) 
dataset
```

```python
#split data to features and target
X = dataset.iloc[:, 1:].as_matrix()
y = dataset.iloc[:, 0].as_matrix()
```

```python
#split data to train and test
X_train = X[:463715]
X_test = X[463715:]
y_train = y[:463715]
y_test = y[463715:]
```

```python
#fit dataset to model 
linReg = LinearRegression(X_train, y_train).fit()
```

```python
#predict the test dataset
preds = linReg.predict(X_test)
```

```python
#calculate MSE with sklearn library
from sklearn.metrics import mean_squared_error
def mse(h, y):
    return mean_squared_error(h, y)
mse(y_test, preds)
```

```python
#calculate RMSE with math sqrt
from math import sqrt
def rmse(h, y):
    return sqrt(mean_squared_error(h, y))
rmse(y_test, preds)
```
